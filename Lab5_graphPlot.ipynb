```r
library(ggplot2)
```

```r
two_grams<- data.frame(files_size_MB=c(0.1,0.4,0.7,1.8,3.5,6.2,11,17),seconds=c(5.20,6.85,7.5,7.87,7.96,7.99,8.40,8.2))
three_grams <- data.frame(files_size_MB=c(0.1,0.4,0.7,1.8,3.5,6.2,11,17),seconds=c(5.35,7.05,7.6,7.8,8.03,8.14,8.45,8.7))
```

```r
ggplot(two_grams, aes(files_size_MB)) +                    
  geom_line(aes(y=two_grams$seconds), colour="green",size=1 ) +  
  geom_line(aes(y=three_grams$seconds), colour="blue", size=1) +
  labs(x="Files size in MB",y="Time taken") 
```

# As seen in the above plot 3 grams (blue) takes more time when compared to 2 grams(green) for any file size
